## Was ist sind Korpora?

- Eine große Sammlung von schriftlichen Texten oder gesprochene Äußerungen 
- basieren auf nachgewiesenen Sprachgebrauch und sind einer Struktur/Ordnung unterworfen

Dabei ist ein Korpus immer nur eine Stichprobe die balanciert und groß und umfassend sein soll. 

__Arten von Korpora:__

- Textkorpora: Zeitungsarikel, Bücher, Blogs
- Gesprochene Sprache: Mitschnitte von Telefonanrufnen, Interviews, Reden usw. 
- Multimodaler Sprachgebrauch: Videoaufzeichnungen von Gesprächen oder Vorträgen 

Beispiele:

- Switchboard Corpus: Spontansprachliche Telefonkonversationen zwischen unbekannten Personen
- Google n-Gram: Aus Google Books: 15 millionen Bücher
- Penn-Treebank: 7 Millionen Wörter aus verschiedenen Ressourcen wie z.B. Wall street Journal

Transkriptionen oder wie man ein Korpus baut:

### Transkription

Sprachdaten liegen in Rohform nur in Audio- oder Videoformaten vor
- Nur bedingt für die automatische Weiterverarbeitung und Analyse geeignet

__Transkription__: Gesprochene Sprache wird in textuelle Repräsentation überführt, ähnlich wie ein Diktat oder eine Abschrift

- Aufwändig, da schneller gesprochen wird als transkribiert werden kann

- Fehleranfällig, da die „Transkriptionsziele“ (Wörter, Pausen, Überlappungen, Betonungen, ...) mitunter undeutlich oder nicht eindeutig identifizierbar sein können

### Annotationen 

- Linguistische Informationen, die einem Korpus hinzugefügt sind

__Warum wird annotiert?__

- Annotationen fügen dem Text Informationen hinzu, die zunächst nicht da sind

Beispiel:
- The book I read yesterday was interesting.
- The book [relRed I read yesterday] was interesting.

- Annotationen erhöhen die Zahl der linguistischen Phänomene, die auffindbar sind
- Annotationen sind für neue Fragestellungen/Aufgaben wiederverwendbar

__Was soll annotiert werden?__

- Aufwendig und teuer dabei sind die immer theoriebasiert und machen Annnahmen
- Gute Annotationssysteme sind nicht einfach zu entwickeln -> Sensitivität, Eindeutigkeit, Verlässlichkeit, Aufwand
- Annotationen spiegeln eine Fragestellung/Aufgabe wider
- Je genereller die Annotation, desto vielfältiger die Nutzungsmöglichkeiten

__Annotationsebenen__

Ebenen oft anhand der Standard-Subdiziplinen der Linguistik
- Phonetik, Morphologie, Syntax, Semantik, Diskurs, Pragmatik
Ebenen reflektieren unterschiedlichen Detailgrad der Annotation
- bspw. grob/fein
Ebenen reflektieren unterschiedliche Modalitäten
- Sprache, Blick, Mimik, Gestik, Biosignale, Ereignisse in der Umgebung, technische Aspekte

__Wie wird annotiert?__

- manuell: durch menschliche AnnotatorInnen
- automatisch: bspw. mit computerlinguistischen Verfahren
- semi-automatisch: Erst automatisch dann manuell korrigiert und fast immer iterativ

Das Verfahren hat Einfluss auf die Art und Häufigkeit von Fehlern
- manuell: eher wenige, aber unsystematische Fehler
- automatisch: ggf. viele, aber systematische Fehler

__Konsistente Annotationen ergeben ein Annotationsschema__

Dieses ist in einem Annotationsmanual dokumentiert

Erforderlich, um Konsistenz in den Annotationen zu erzielen:
- Definitionen der Annotationskategorien
- Vokabular, Listen von Labels
- Erklärungen zum Annotationsprozess (Kriterien)

__Qualität von Annotationen__

Wichtiges Kriterium für Annotation: 
- Konsistenz und Reliabilität 

wie kann man das messen?

- Eine Stichprobe der Daten (ca. 10%) wird von Annotatoren unabhängig voneinander annotiert (Cross validation)
- Bestimmung des Agreements zwischen den Probe-Annotationen.

Es gibt verschiedene mathemaische Maße, die man verwenden kann. Das gebräuchlichste ist Cohen’s Kappa:

Formel: $k = \frac{p_0-p_e}{1-p_e}$

dabei gilt:

$p_0$ = Relative beobachtete Übereinstimmung
$p_e$ = Erwartete Wahrscheinlichkeit für eine Zufallsübereinstimmung

### Morphologie

Wörter werden aus Morphemen gebildet

- Morphem: kleinste bedeutungstragende Einheit auf der Ebene des Sprachsystems. 

__Arten von Wortbildung__

Flexion: Ändert bei gleichem Stamm die Werte von Merkmalen und ggf. die Form
- Folie [Num: sg] → Folie-n [Num: pl]

Derivation: Bildet aus einem Stamm durch Affigierung einen neuen Stamm
- Läuf:er, Un:mensch

Komposition: Wortstamm wird aus zwei oder mehreren Stämmen anderer lexikalischer Wörter zusammengesetzt
- Feuer.wehr, Feuer.wehr.wache, feuer.rot

Konversion: Der Stamm wird weiterverwendet, es ändert sich jedoch die Wortklasse
- einkaufen [V] → das Einkaufen [N]

__Arten der Auszeichnung__

Die Wortbildung kann auf unterschiedliche Art ausgezeichnet werden.
Durch Hinzufügung von Metadaten (Feature structures / Merkmalsstrukturen)
- Folien [Klasse: N, Num: pl, Kas: nom, Gen: fem]

Auszeichnung
- Folie+N+Pl+Nom+Fem
