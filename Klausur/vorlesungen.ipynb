## Was ist sind Korpora?

- Eine große Sammlung von schriftlichen Texten oder gesprochene Äußerungen 
- basieren auf nachgewiesenen Sprachgebrauch und sind einer Struktur/Ordnung unterworfen

Dabei ist ein Korpus immer nur eine Stichprobe die balanciert und groß und umfassend sein soll. 

__Arten von Korpora:__

- Textkorpora: Zeitungsarikel, Bücher, Blogs
- Gesprochene Sprache: Mitschnitte von Telefonanrufnen, Interviews, Reden usw. 
- Multimodaler Sprachgebrauch: Videoaufzeichnungen von Gesprächen oder Vorträgen 

Beispiele:

- Switchboard Corpus: Spontansprachliche Telefonkonversationen zwischen unbekannten Personen
- Google n-Gram: Aus Google Books: 15 millionen Bücher
- Penn-Treebank: 7 Millionen Wörter aus verschiedenen Ressourcen wie z.B. Wall street Journal

### Annotationen 

- Linguistische Informationen, die einem Korpus hinzugefügt sind

__Warum wird annotiert?__

- Annotationen fügen dem Text Informationen hinzu, die zunächst nicht da sind

Beispiel:
- The book I read yesterday was interesting.
- The book [relRed I read yesterday] was interesting.

- Annotationen erhöhen die Zahl der linguistischen Phänomene, die auffindbar sind
- Annotationen sind für neue Fragestellungen/Aufgaben wiederverwendbar

__Was soll annotiert werden?__

- Aufwendig und teuer dabei sind die immer theoriebasiert und machen Annnahmen
- Gute Annotationssysteme sind nicht einfach zu entwickeln -> Sensitivität, Eindeutigkeit, Verlässlichkeit, Aufwand
- Annotationen spiegeln eine Fragestellung/Aufgabe wider
- Je genereller die Annotation, desto vielfältiger die Nutzungsmöglichkeiten

__Annotationsebenen__

Ebenen oft anhand der Standard-Subdiziplinen der Linguistik
- Phonetik, Morphologie, Syntax, Semantik, Diskurs, Pragmatik
Ebenen reflektieren unterschiedlichen Detailgrad der Annotation
- bspw. grob/fein
Ebenen reflektieren unterschiedliche Modalitäten
- Sprache, Blick, Mimik, Gestik, Biosignale, Ereignisse in der Umgebung, technische Aspekte

__Wie wird annotiert?__

- manuell: durch menschliche AnnotatorInnen
- automatisch: bspw. mit computerlinguistischen Verfahren
- semi-automatisch: Erst automatisch dann manuell korrigiert und fast immer iterativ

Das Verfahren hat Einfluss auf die Art und Häufigkeit von Fehlern
- manuell: eher wenige, aber unsystematische Fehler
- automatisch: ggf. viele, aber systematische Fehler

__Konsistente Annotationen ergeben ein Annotationsschema__

Dieses ist in einem Annotationsmanual dokumentiert

Erforderlich, um Konsistenz in den Annotationen zu erzielen:
- Definitionen der Annotationskategorien
- Vokabular, Listen von Labels
- Erklärungen zum Annotationsprozess (Kriterien)

__Qualität von Annotationen__

Wichtiges Kriterium für Annotation: 
- Konsistenz und Reliabilität

### Morphologie

Wörter werden aus Morphemen gebildet

- Morphem: kleinste bedeutungstragende Einheit auf der Ebene des Sprachsystems. 

__Arten von Wortbildung__

Flexion: Ändert bei gleichem Stamm die Werte von Merkmalen und ggf. die Form
- Folie [Num: sg] → Folie-n [Num: pl]

Derivation: Bildet aus einem Stamm durch Affigierung einen neuen Stamm
- Läuf:er, Un:mensch

Komposition: Wortstamm wird aus zwei oder mehreren Stämmen anderer lexikalischer Wörter zusammengesetzt
- Feuer.wehr, Feuer.wehr.wache, feuer.rot

Konversion: Der Stamm wird weiterverwendet, es ändert sich jedoch die Wortklasse
- einkaufen [V] → das Einkaufen [N]

__Arten der Auszeichnung__

Die Wortbildung kann auf unterschiedliche Art ausgezeichnet werden.
Durch Hinzufügung von Metadaten (Feature structures / Merkmalsstrukturen)
- Folien [Klasse: N, Num: pl, Kas: nom, Gen: fem]

Auszeichnung
- Folie+N+Pl+Nom+Fem

Dabei hat jede Sprache seine eigenen Flexion usw. und Komplexität

---

### Parsing

Parsing means taking an input and identifying/producing some sort of linguistic structure for it.

__Morphologisches Parsing__

Abbildung von Wörtern auf ihre morphologische Struktur:
- cat → cat+N+Sg; cats → cat+N+Pl
- Folie → Folie+N+Sg+Akk+Fem

Für das Morphologische Parsing werden folgende Dinge benötigt:
1. Lexikon
    - Liste alle Wortstämme und Affixe inkl. Basisinformationen
2. Morphotaktik
    - Regeln/Modell der Ordnung von Morphemklassen – welche Klassen von Morphemen folgen anderen Klassen von Morphemen innerhalb eines Wortes
3. Orthographische Regeln
    - Modellierung der morphologisch bedingten Änderungen in einem Wort

__Lexikon__ 

Ein Lexikon könnte einfach eine Auflistung aller Wörter einer Sprache sein. Ein solches Vollformenlexikon ist je nach Sprache aber nicht besonders effizient oder sogar unmöglich zu konstruieren.

Stattdessen sollten Wortstämme und Affixe zusammen mit der Morphotaktik (Regeln der zulässigen Anwendung) repräsentiert werden
- Angabe von Primitiven plus Produktions-/Ableitungsregeln
- Modellierung als endlicher Automat (Finite State Machine/Automaton) oder als formale Grammatik

__Finite State Transducer__

Transducer bilden zwei Repräsentationen schrittweise aufeinander ab.

Finite State Transducer (FST) sind endliche Automaten, die zwei Mengen von Symbolen zustandsbasiert aufeinander abbilden
- Automat mit zwei „Bändern“, der Paare von Strings erkennen oder generieren kann
- FST definiert Relationen zwischen Mengen von Strings
- FST kann als Übersetzer genutzt werden

### Konstituenz
---

Gruppen von Wörtern verhalten sich wie eine Einheit -> Konstituente/Phrase

Beispiel: Nominalphrase (NP):
- Harald der Hase; drei Gruppen aus Bielefeld; rote Zahnbürsten des Königs die benutzt waren;

__Welche Wortfolge ist eine Konstituente?__

1. Substitutionstest
Kann man eine Wortfolge in einem Satz gegen eine andere Wortfolge so austauschen, dass wieder ein akzeptabler Satz entsteht, so ist das ein Indiz dafür, dass die beiden Wortfolgen Konstituenten bilden.
    - Er kennt [den Mann]. → Er kennt [die Frau].
    - Er versucht [das Buch zu lesen]. → Er versucht [der Frau das Buch zu geben].

2. Pronominalisierungstest
Alles, worauf man sich mit einem Pronomen beziehen kann, ist eine Konstituente.
    - [Der Mann] schläft. → [Er] schläft.  
    - Peter versucht [das Buch zu lesen]. → Klaus versucht [das auch].

3. Fragetest
Was sich erfragen lässt, ist eine Konstituente.
    - [Der Mann] arbeitet. → Wer arbeitet? [Der Mann]
    - Er versucht [das Buch zu lesen]. → Was versucht er? [Das Buch zu lesen]

4. Permutationstest
Wenn Wortfolgen ohne Beeinträchtigung der Akzeptabilität des Satzes verschoben bzw. umgestellt werden können, ist das ein Indiz dafür, dass sie eine Konstituente bilden.
    - dass keiner [diese Frau] kennt → dass [diese Frau] keiner kennt
    - Er hat noch nicht [das Buch zu lesen] versucht. → Er hat [das Buch zu lesen] noch nicht
versucht. → Er hat noch nicht versucht [das Buch zu lesen].

5. Voranstellungstest
Die Voranstellbarkeit einer Wortfolge ist als starkes Indiz für deren Konstituentenstatus zu werten (Es steht genau eine Konstituente vor dem finiten Verb).
    - [Alle Studenten] lesen während der vorlesungsfreien Zeit Bücher.
    - [Bücher] lesen alle Studenten während der vorlesungsfreien Zeit.

6. Koordinationstest
Lassen sich Wortfolgen koordinieren, so ist das ein Indiz dafür, dass die koordinierten Wortfolgen jeweils Konstituenten sind.
    - [Der Mann] und [die Frau] arbeiten.
    - Er hat versucht, [das Buch zu lesen] und [es dann unauffällig verschwinden zu lassen].

__Konstituententests sind nur Indizien.__

Als Kopf einer Konstituente/Phrase bezeichnet man das Element, welches die wichtigste Eigenschaft bestimmt.
- Träumt dieser Mann?
- Erwartet er diesen Mann?

Der Kopf steuert den Aufbau der Phrase
- welche Elemente erforderlich sind
- welche Merkmale der Kontext der Phrase hat (Projektion)

### Parsing

__Syntaktisches Parsing__

Parsing: Einen Eingabestring erkennen und eine (linguistische) Struktur zuweisen

Parsing basiert auf Suche:
- Finde alle Bäume, die S als Startsymbol haben und die genau den Eingabestring abdecken
- Verschiedene Suchstrategien

__Kontextfreie Grammatik__

Kennt man schon aus GTI 
- N: Eine Menge von Nicht-Terminalsymbolen/ Variablen
- Σ: Eine Menge von Terminalsymbolen
- R: Eine Menge von Regeln der From A → β,
mit A ∈ N und β ∈ (Σ ∪ N)*
- S: Ein Startsymbol S ∈ N.
Die Grammatik heisst kontextfrei, weil auf der linken Seite jeder Regel immer nur eine Variable steht
→ die Regelanwendung ist unabhängig vom die Variable umgebenden Kontext.

__Top-down Parsing__

Suche, ausgehend vom Startsymbol S, einen Parse Tree der zu den Blättern (d.h. Wörtern der Eingabe) passt.

__Bottom-up Parsing__

Baue, ausgehend von den Blättern (d.h. Wörtern der Eingabe), einen Parse Tree auf, der im Startsymbol S wurzelt.

__Top-down vs Bottom-up__

Beide Strategien haben Vor- und Nachteile

Top-down
- Es werden keine Bäume exploriert, die nicht in S resultieren können.
- Es werden viele Bäume exploriert, die nicht zur Eingabe passen.
Bottom-up
- Es werden viele Bäume generiert, die nicht in S resultieren können
- Es werden keine Bäume exploriert, die inkompatibel mit der Eingabe sind.

### CKY-Parsing

Syntaktische Ambiguität 

Strukturelle Ambiguitäten existieren auf allen Ebenen und machen das Parsing komplex.
- Attachment-Ambiguität: I shot an elephant in my pyjamas. (How he got into my pyjamas I don’t know.)
- Coordination-Ambiguität: old men and women
    - [old [men and women]]
    - [[old men] [and] [women]]

viele alternative Parses müssen gespeichert und verglichen werden

Parsing basierend auf dem Prinzip der dynamischen Programmierung:
- Bestimmung und Berechnung aller Teilprobleme, die gelöst werden müssen
- Geschickte Lösung der Teilproblemen ➔ speichern in Tabelle von Teillösungen
- Wiederverwendung von Teillösungen, wo immer effizient möglich

CKY-Parsing: Speicher alle Teilbäume für die in der Eingabe gefundenen Konstituenten
- Teilbäume werden nur einmal gesucht und später für das Parsen der Konstituente
wiederverwendet
- Berücksichtigt das Ambiguitätsproblem, da alle Parses implizit gespeichert werden
- Voraussetzung: Grammatik liegt in Chomsky-Normalform vor

__Chromsky Normalform__

Eine kontextfreie Grammatik ist in Chomsky-Normalform (CNF), gdw.
- sie ε-frei ist, und
- jede ihrer Produktionsregeln die Form A → B C oder A → a hat.

Die rechte Seite hat also entweder zwei Nicht-Terminalsymbole oder ein Terminalsymbol ➔ alle Regeln bis auf die Prä-Terminale erzeugen Binärbäume.

Gilt: Jede kontextfreie Grammatik kann in Chomsky-Normalform gebracht werden. 
    - Aus A → B C D wird A → X D; X → B C

__WICHTIG: CKY-PARSING VERSTEHEN UND ANWENDEN!__

Dabei findet CKY-Parsing antworten auf: 

1. Recognition-Problem (Wortproblem):
- Gibt es ein Start-Symbol in Tabellenzelle [0, N]? — CKY liefert direkt und effizient die Antwort.

2. Parsing-Problem (eindeutige Zuweisung einer linguistischen Struktur):
- Vom Start-Symbol in Zelle [0, N] ausgehend rekursiv die Komponenten- Konstituenten aus der Tabelle auslesen — aufgrund der Mehrdeutigkeiten (Mehrfachbelegung von Zellen der Tabelle) nicht eindeutig möglich.

### Repräsentation von Bedeutung

Annahme: Die Bedeutung linguistischer Ausdrücke lässt sich in formalen Strukturen (Repräsentationen) „einfangen“.
- Bindeglied zwischen linguistischer Eingabe und nicht-linguistischem Weltwissen, welches für die (linguistische) Aufgabe benötigt wird.
- Bedeutungsrepräsentationen analog zu anderen Arten von Wissensrepräsentationen (vgl. Vorlesung GKK)

Die Erzeugung und Zuweisung von Bedeutungsrepräsentationen zu linguistischen Strukturen nennt man semantische Analyse / semantisches Parsing

__Perspektiven__

Zwei Perspektiven auf Repräsentation
1. Repräsentation des Inhalts einer linguistischen Struktur (Wort, Satz, Phrase)
2. Repräsentation der Bedeutung als ein Zustand in einer Welt

-> Ermöglicht eine Verbindung von Inhalt und Bedeutung

__Desiderata__

Verifizierbarkeit: Es muss möglich sein, die Wahrheit/Gültigkeit der Repräsentation zu überprüfen
- z.B. durch Abgleich der Repräsentation mit dem Wissen in einer Wissensbasis KB

Eindeutigkeit: Die Repräsentation soll nicht ambig sein. Referentielle, semantische, funktionale Eindeutigkeit (vgl. VL GKK)

Kanonische Form:Verschiedene Sätze können (ungefähr) die gleiche Bedeutung haben
- Does Maharani have vegetarian dishes?
- Do they have vegetarian food at Maharani?
- Are vegetarian dishes served at Maharani?

Verschiedene Bedeutungsrepräsentationen dieser Fragen erschweren abgleich mit der Wissensbasis.
Die Antworten auf all diese Fragen sollten jedoch gleich sein → es ist wünschenswert, dass sie auf die gleiche Bedeutungsrepräsentation abbilden.

Inferenz und Variablen:
- Beispiel: Can vegetarians eat at Maharani?
    - Hier wäre es falsch, die gleiche kanonische Form anzunehmen
    - Einfaches Matching mit der Wissensbasis ist ggf. nicht ausreichend
    → erfordert Inferenz um mittels Weltwissen Fakten zu schlussfolgern, die nicht explizit in der Wissensbasis repräsentiert sind

Ausdrucksstärke/Expressivität: Idealerweise sollte die Bedeutung jeglicher sprachlicher Äußerungen repräsentiert werden können. 
- Beispiel: Prädikatenlogik erster Ordnung (PL1 / FOL) kann ein signifikantes Fragment/eine Domäne repräsentieren, aber nicht alle in Sprache ausdrückbare Inhalte und Bedeutung

Repräsentationsschema: Kann Objekte, Objekteigenschaften, Relationen zwischen Objekten repräsentieren
Modell eines Weltzustands: Ausdrücke in der Bedeutungsrepräsentationssprache können systematisch auf Elemente des Modells abgebildet werden. Wenn das Modell akkurat ist (d.h. der Welt tatsächlich entspricht), haben wir eine funktionierende Brücke zwischen Sprache und Welt

Jedes Element der Bedeutungsrepräsentation benötigt eine Denotation im Modell.

Domäne eines Modells: Menge der Objekte im Zustand der Welt
- Objekte denotieren Elemente der Domäne
- Eigenschaften denotieren Mengen von Elementen der Domäne
- Relationen denotieren Mengen von Tupeln von Elementen der Domäne

Interpretation: Eine Funktion die von dem Vokabular der Bedeutungsrepräsentation auf die Denotationen im Modell abbildet.

### Lambda Abstraktion und Reduktion

Lambda-Ausdruck: eine freie Variable x wird durch λ-Abstraktion gebunden 
    - λx.P(x)
λ-Applikation dieses Ausdrucks auf eine Konstante A (kann auf jede FOL-Formel appliziert werden)
    - λx.P(x)(A)
λ-Reduktion erfolgt dann, indem jedes Vorkommen von x in P(x) an A gebunden wird
    P(A)

Lambda-Ausdrücke können auf FOL-Ausdrücke angewendet werden und ergeben neue FOL-Ausdrücke, bei denen die Parametervariablen an spezifische Terme gebunden werden:
- λx.P(x)(A) → P(A)
- λx.λy.Near(x,y)(Maharani) → λy.Near(Maharani, y)
- λy.Near(Maharani, y)(ICSI) → Near(Maharani, ICSI)

### Syntaxgetriebene semantische Analyse

Prinzip der Kompositionalität 

Idee: Die Bedeutung eines Satzes kann aus den Bedeutungen der einzelnen Teile des Satzes konstruiert werden.

- die Bedeutung eines Satzes basiert (partiell) auf seinen Bestandteilen und seiner syntaktischen Struktur.
- um die Bedeutung eines Satzes zu „komponieren“, lässt man sich durch die syntaktischen Komponenten und die durch die Grammatik zur Verfügung gestellten Relationen leiten.

__Was ist eigentlich die Bedeutung eines Wortes?__

Lässt sie sich im Wörterbuch finden?
- normativ, formalisiert, definitorisch
Eine alternative Sicht:
- Die Bedeutung eines Wortes ist sein Gebrauch in der Sprache

Distributional Hypothesis:
Wörter sind in ihrer Bedeutung umso ähnlicher, je mehr sie in gleichen Verteilungen auftreten und je mehr sie in ähnlichen Kontexten benutzt werden.
Basis für den Ansatz der Distributional (statistical) Semantics: Repräsentation von Wortbedeutung als statistische Information über die Verteilung des Gebrauchs des Wortes in Sprachdaten

__Wortbedeutung lässt sich in Zahlen fassen lassen__

Beispiel: Affektive Zustände lassen sich anhand von drei Achsen (Valenz, Erregung, Dominanz) charakterisieren.
Wörter haben affektive Bedeutungen und können in diesem Raum positioniert werden:
- happy/sad: hohe/niedrige Valenz
- excited/calm: hohe/niedrige Erregung
- important/awe: hohe/niedrige Dominanz

### Diskurs

__Was ist ein Diskurs?__

Sprache besteht nicht aus einzelnen, für sich stehenden Wörtern oder Sätzen.
Kollokierte Sätze treten in strukturierten, kohärenten Gruppen auf und bilden
einen Diskurs.

---

### Kohärenz

→ Sätze bilden einen kohärenten Diskurs, wenn sie inhaltlich zueinander in Verbindung oder in einer Beziehung stehen.

Beispiel:

a. John went to his favorite music store to buy a piano.
b. He had frequented the store for many years.
c. He was excited that he could finally buy a piano.

→ Sätze bilden einen kohärenten Diskurs, wenn sie Beziehungen zwischen den Dingen um die es geht aufzeigen, sie einführen, und ihnen fokussiert folgen.

__Kohäsion__

a. Before winter I built a chimney, and shingled the sides of my house ... I have thus a tight shingled and plastered house ...
b. Peel, core and slice the pears and the apples. Add the fruit to the skillet.
c. The Woodhouses were first in consequence there. All looked up to them.

→ Kohäsion ist die Verwendung linguistischer Merkmale („devices“) um textuelle Einheiten miteinander zu verbinden, hier z.B. lexikalische Kohäsion. 

__Kohärenz vs Kohäsion__

Kohäsion: Bezeichnet die erkennbare Verbundenheit zweier Einheiten durch formale linguistische Oberflächenmerkmale (grammatikalisch, lexikalisch)
- Eine Art „Kleber“, der zwei Einheiten zu einer einzelnen Einheit gruppiert
- Geknüpft an explizite, im Text vorkommende Elemente.

Kohärenz: Semantische Bedeutungsrelationen zwischen zwei Einheiten, erklären wie die Bedeutung verschiedener Einheiten kombiniert werden um eine gemeinsame Diskursbedeutung einer größeren Einheit zu erlangen.
- Generelles Prinzip der Interpretation von Sprache im Kontext
- Eher inhaltlicher Begriff, bezogen auf den Sinnzusammenhang eines Textes.

Was ist der Zusammenhang zwischen Kohäsion und Kohärenz?
Wird beides benötigt? Sind das unabhängige Dimensionen, oder folgt aus dem einen das andere?

Beispiele mit Kombinationen:
1. Sie trug den Kopf im Kleid vor sich. Wie ein Känguru sah sie aus. (W. Borchert: „Das Känguru“)
- kohäsiv und kohärent
2. Judith trug den Kopf im Kleid vor sich, welches sie eigens wie einen Beutel hochgefaltet hatte.Wie ein Känguruh sah sie aus.
- kohäsiver und kohärenter als 1.
3. Die Sonne schien blutrot über dem Horizont. Wie ein Känguru sah sie aus.
- kohäsiv und nicht kohärent
4. Die Posten dösen. Wie ein Känguruh sah sie aus.
- nicht kohäsiv und nicht kohärent

__Zusammenahng und zusammenhalt__

Was ist der Zusammenhang? Wird beides benötigt? Folgt aus dem einen das andere?

Kohäsion und Kohärenz sind keine identischen, aber auch keine unabhängigen Dimensionen!
- Diskurse (z.B. Texte) sollten immer kohärent sein – inkohärent ≈ unverständlich.
- Kohäsion erleichtert es, beabsichtigte Kohärenz zu erkennen
- Normale Texte sind meist sowohl kohäsiv als auch kohärent.

### Diskursrelationen

__Typen__

Result:
- The Tin Woodman was caught in therain. His joints rusted.
Elaboration:
- Dorothy was from Kansas. She lived in the mist of the great Kansas prairies.
Explanation:
- John hid Bill’s car keys. He was drunk
Parallel:
- The Scarecrow wanted some brains. The Tin Woodman wanted a heart
Occasion:
- Dorothy picked up the oil-can. She oiled the Tin Woodman’s joints.

__Rhetorical Structure Theory__

Eine Menge wichtiger Diskursrelationen erfasst die Rhetorical Structure Theory:
- Definiert Relationen zwischen (zumeist) zwei Textteilen, bezeichnet als
    - Nucleus: zentrale Einheit,
    - Satellite: weniger zentrale Einheit, kann i.A. nur zusammen mit dem Nucleus verstanden werden
- Grafische Darstellung durch Schemata
    - geschwungene Pfeile: Relationen vom Satellite zum Nucleus
    - gerade Linien: definieren Textteile

### Automatische Diskursanalyse und Generierung

__Discourse Parsing__

Beim Discourse Parsing wird einem Text eine hierarchische Struktur von Diskusrelationen zugeordnet.

Shallow Discourse-Parsing: Oft wird Diskursstruktur im Text durch Cue-Phrasen, sogenannten Diskursmarkern, markiert: „John hid Bill’s car keys because he was drunk.“

Beim Discourse Parsing wird einem Text eine hierarchische Struktur von Diskusrelationen zugeordnet.

Shallow Discourse-Parsing
- Oft wird Diskursstruktur im Text durch Cue-Phrasen, sogenannten Diskursmarkern, markiert: „John hid Bill’s car keys because he was drunk“
- Die Verwendung von Cue-Phrases ist allerdings nicht eindeutig, da sie auch andere Funktionen im Text haben können.
- Identifikation von Cue-Phrasen, Segmentierung des Textes mittels der Cue-Phrasen, Zuordnung von Diskursrelationen zu konsekutiven Segmenten
-> Klassifikationsprobleme

__Diskursgenerierung__

Die Definition von RST-Relationen ermöglicht prinzipiell die Generierung von Texten mit KI-Planungsoperatoren
- Constraints on Nucleus/Satellite → Preconditions für die Anwendung eines Operators
- Effects → Postconditions der Anwendung eines Operators

Alternativ können auch direkt Textteile mittels Cue-Phrases verbunden werden.

Top-down Generierung: Dieser Ansatz eignet sich gut für Domänen, in denen vorhersehbar ist, welche Art von Daten vorliegen.
Bottom-up Generierung: Dieser Ansatz eignet sich für Domänen, in denen nicht vorhersehbar ist, welche Messages verbalisiert werden sollen.

__Aktuelle Ansätze und Probleme__

- Transformer-based models perform better on linguistically different intra-
sentential relations than they do on inter-sentential relations.
- Parsers struggle to identify implicit relations from less frequent classes.
- Models fail to generalize to both in-domain and out-of-domain data, and different errors are seen for different domains.
- More context than the two arguments is needed to determine the correct discourse relation in many cases.
- Attachment issues tend to occur throughout the RST parse tree.

Was heißt verstehen? Seit der Antike bis heute: genau die Nachricht (Bedeutung) aus einer Äußerung extrahieren, die Sprecher*in zu übertragen intendierte

__Starkes Verstehen__

Dies führt zu einem starken Konzept von Verstehen – einem theoretischen Ideal, welches zwei Annahmen macht:
- Konformität der Repräsentation (wir haben alle die gleichen Konzepte)
- Konformität der Intersubjektivität (wir nutzen alle die gleichen Worte für diese
Konzepte)

->  Diese Annahmen sind zu stark — Sprache als Kommunikationsmittel ist eigentlich ungenügend, um Ideen und Gedanken in diesem Sinne exakt zu kommunizieren.

__Schwaches Verstehen__

Diese Probleme könnten in gewisser Hinsicht vereinfacht werden, da Menschen...
- Konzepte in einer gemeinsamen natürlichen und kulturellen Umwelt erwerben, eine Ähnlichkeit zwischen Konzepten also wahrscheinlich ist
- Wortbedeutung in sozialen Interaktionen erlernen, in denen sie gemeinsame Referenzrahmen und geteilte Aufmerksamkeit haben (konventionalisierte Wortbedeutungen verbreiten sich in Sprachgemeinschaften)

-> Schwaches Verstehen: eine Nachricht (Bedeutung) aus einer Äußerung extrahieren, die ähnlich zu der ist, die die Sprecher*in zu übertragen intendierte

Wörtliche Bedeutung != Bedeutung des Sprechers

__Verstehen und Bedeutung__

Das Problem von speaker meaning vs sentence meaning
- Zentrale Annahme für das Konzept des Verstehens von wörtlicher Bedeutung ist, dass Kommunikation auf Konvention aufbaut (Übereinkommen bzgl. linguistischer Struktur).
- Speaker meaning korrespondiert aber nicht immer und nicht komplett zur wörtlichen Bedeutung!

-> Wörtliche Bedeutung ist nur eine Informationsquelle
-> Verstehen ist ein viel schwierigeres (vielleicht unmögliches?) Inferenzproblem

__Probleme des Verstehens__

1. erstehen ist subjektiv: Im Allgemeinen ist es nicht einfach möglich für individuelle Gesprächsteilnehmer herauszufinden, ob XS ≈ XL
2. Gesprächspartner können sich bewusst werden, das es ein Verstehensproblem gibt → Nichtverstehen
3. Gesprächspartner ist ggf. nicht bewusst, dass es ein Verstehensproblem gibt → Missverstehen

Nicht- und Missverstehen sind die Standardfälle von Kommunikation:
- Gesprächsteilnehmer gehen nicht von perfektem Verstehen aus, sie sind immer auch auf Nichtverstehen und Missverständnisse vorbereitet
- Sie können zuversichtlich sein, dass diese Probleme durch Mittel sozialer Interaktion entdeckt und behoben werden können
- In linguistischer Interaktion geht es nicht um einen kognitiven Endzustand von Verstehen, sondern um das Erreichen von Verstehen (“coming to an understanding”)
- Verstehen ist dabei stets subjektiv von den Kommunikationspartnern bewertet

Verschiedene Prozesse sind involviert:
- Es müssen Repräsentationen über die Überzeugungen (beliefs) und Ansichten (attitudes) der anderen aufgebaut und, gegeben neuer Informationen, aktuell gehalten werden
- Es müssen meta-kommunikative Informationen über die eigenen Überzeugungen und Ansichten bereitgestellt werden.
- Kommunikative Akte müssen vor dem Hintergrund der vom Adressaten vermutlich gehaltenen Überzeugungen und Ansichten gestaltet werden

-> Bedeutung wird oft schrittweise, über mehrere Turns hinweg kommuniziert und graduell verbessert, bis an einem gewissen Punkt beide Gesprächspartner gemeinsam glauben, dass ein Verstehen erreicht wurde, welches „für die aktuellen Zwecke ausreicht“
(siehe VL zu „Grounding“)

__Performative und Konstantive Äußerungen__

(S1) Ich taufe dieses Schiff auf den Namen „Wüste Walli“.
(S2) Ich wette fünf Euro, dass Sansar den 48. Hermannslauf gewinnt. 
(S3) Mein Kollege heißt Christopher und mein Mentor heißt Stefan. 
(S4) Ich wohne und arbeite in einer kleineren Großstadt in Europa.

(S1) und (S2) sind deklarative Sätze
- Durch das Äußern derartiger Sätze wird etwas getan (es wird ein Akt „performed“)
- Gekennzeichnet durch performative Verben (der Form 1. Person-Singular-Präsenz- Indikativ-Aktiv), die das Getane explizit machen

-> Performative Äußerungen sind nicht wahr oder falsch

(S3) und (S4) sind assertative Sätze
- Durch das Äußern derartiger Sätze wird etwas ausgesagt, das der Fall ist

-> Konstative Äußerunge sind wahr oder falsch

Performative können aber misslingen oder „missglücken“
- Problem bei (S2): Ich kann nicht wetten, denn der 48. Hermannslauf fand 2019 statt.

Für Performative gibt es eine Reihe von Gelingensbedingungen
- (A1) Es existiert eine konventionalisierte Prozedur mit einem konventionalisierten Effekt
- (A2) Personen und Umstände müssen entsprechend der Prozedur angemessen sein
- Die Prozedur muss (B1) korrekt und (B2) vollständig ausgeführt werden.

Konstantive Äußerungen können auch mislingen, wenn es nicht genau zutrifft oder falsch ist.

__Komponenten (Akte) einer Äußerung__

Eine Äusserung kann als sogenannte Lokution betrachtet werden, die mehrere Handlungsaspekte („Akte“) beinhaltet.

Lokutionärer Akt
- Der Akt „etwas zu sagen“
- Kann weiter unterteilt werden in andere Akte

Illokutionärer Akt
- Der Akt „durch etwas zu sagen etwas zu tun“
    - beschuldigen, Fragen stellen oder beantworten, entschuldigen, verantwortlich machen, informieren, bestellen, sicher stellen, warnen, eine Intention ankündigen, einen Termin vereinbaren, eine Beschreibung geben, versprechen, feststellen, ...
- Jeder lokutionäre Akt ist gleichzeitig auch ein illokutionärer Akt.
- Illokutionäre Akte entsprechen Konventionen und haben konventionalisierte Kräfte
(die illokutionäre Kraft*), die gewisse Effekte verursacht.

Perlokutionärer Akt
- Der Akt „durch etwas zu sagen andere dazu zu bringen etwas zu tun“
    - überreden, alarmieren, überzeugen, abschrecken, überraschen, jemanden dazu bringen etwas zu tun, ...
- Jeder illokutionäre Akt ist gleichzeitig auch ein perlokutionärer Akt
- Perlokutionäre Akte haben Effekte auf die Gefühle, Gedanken und Handlungen der Adressaten und haben somit psychologische und/oder behaviorale Konsequenzen
- Der Effekt eines perlokutionären Aktes ist nicht konventionalisiert.

__Effekte Illokutionärer Akte__

Eine notwendige Bedingung für den Erfolg eines illokutionären Aktes ist das Uptake durch den Adressaten
- muss sicher gestellt werden, um den Kerneffekt eines illokutionären Aktes zu erzielen

Illokutionäre Akte haben drei Klassen von Effekten:
1. Sichern von Uptake → Verständnis der Bedeutung und der Kraft der Lokution
2. Produktion eines konventionalisieren Effekts
3. Einladung von Reaktionen

Referenzielle Ausdrücke: 
- Any expression which serves to identify any thing, process, event, action, or any other kind of individual or particular I shall call a referring expression.

### Prinzip der Kooperativität (Was zeichnet eine gute Referenz aus?)

1. Maxime der Quantität
- Mache deinen Gesprächsbeitrag mindestens so informativ, wie es für den anerkannten Zweck des Gesprächs nötig ist.
- Mache deinen Beitrag nicht informativer, als es für den anerkannten Zweck des Gesprächs nötig ist.

2. Maxime der Qualität
- Versuche einen Gesprächsbeitrag zu liefern, der wahr ist.
- Sage nichts, wovon du glaubst, dass es falsch ist.
- Sage nichts, wofür du keine hinreichenden Anhaltspunkte hast.

3. Maxime der Relevanz
- Sage nichts, was nicht zum Thema gehört, wechsle nicht das Thema.
- Beachte den Gesprächskontext vorangegangener Kommunikation und das Vorwissen deines Kommunikationspartners.

4. Maxime des Stils (der Klarheit)
- Vermeide Unklarheit
- Vermeide Mehrdeutigkeit.
- Vermeide unnötige Weitschweifigkeit.
- Vermeide Ungeordnetheit.


Full Brevity Algorithmus:
- Generierung der kürzest möglichen unterscheidenden Beschreibung
- Probleme: 
    - dieser Algorithmus ist NP-hart
    - Menschen produzieren regelmäßig Beschreibungen, die nicht minimal sind. D.h., sie überspezifizieren.

Greedy Heristic Algorithmus
- Eine Approximation des „full brevity“-Algorithmus:
    - Wähle iterativ diejenige Eigenschaft, die die größte Menge von noch vorhandenen Distraktoren eliminiert („greedy heuristic“).
    - Erweitert die Beschreibung in jedem Schritt um die Eigenschaft mit der größten diskriminatorischen Macht.
- Problem: Generiert nicht immer minimale unterscheidbare Beschreibungen

Incremental Algorithmus
- Erweiterung des „greedy heuristic“-Algorithmus um die Idee, dass Sprecher manche Eigenschaften bei der Beschreibung präferieren
- Intuition: dispräferiertere Attribute werden erst in Betracht gezogen, wenn andere keine unterscheidende Beschreibung ermöglichen.
- Beschreibungen werden inkrementell angereichert, die Attribute werden jedoch in einer festen domänenspezifischen Reihenfolge überprüft.
- Einfach zu implementieren, effizient (kein Backtracking).

### Common Ground and Mutual Knowledge

Subjektives Wissen und "Common Ground"

Wichtige Unterscheidung: 
- geteiltes Wissen / shared knowledge — zwei Personen haben zumindest teilweise identisches Wissen über eine Domäne (z.B. ein Physiker aus Japan und ein Physiker aus Dänemark)
- gemeinsames Wissen / mutual knowledge — zwei Personen haben zumindest teilweise identisches Wissen über eine Domäne und wissen das voneinander

Abschätzen von Mutual Knowledge
- Abschneideheuristik: es reicht shared knowledge2/3/4/5 statt mutual knowledge, permanente Schwächung von Annahme 1
- Kopräsenz-Heuristiken:
    - was sind Belege dafür, dass Bedingungen (1), (2), (3), ... gelten?
    - es werden Belege benötigt, aus denen (1), (2), (3), ... inferiert werden können
    - die Idee ist, man muss nicht alles wissen, aber (bei Bedarf) herleiten können

Belege durch Kopräsenz erfordern die folgenden Hilfsannahmen:
- Gleichzeitigkeit
- Aufmerksamkeit
- Rationalität

Inferenz von Mutual Knowledge
- A and B mutually know that p if and only if some state of affairs G holds such that:
    - A and B have reason to believe that G holds.
    - G indicates to A and B that each has reason to believe that G holds.
    - G indicates to A and B that p.
- G ist eine shared basis für mutual knowledge that p.
- Keine der unendlichen Bedingungen für mutual knowledge müssen überprüft werden.

### Grounding als Prozess

Contributions: Sind nicht Handlungen einzelner Gesprächspartner, sondern gemeinsame Handlungen beider 
Contributions bestehen aus zwei Phasen
1. Presentation
- Einer der Gesprächspartner (S) macht eine Äußerung mit der sie versucht den Inhalt den sie zum Diskurs beitragen will zu präsentieren
2. Acceptance
- Der andere Gesprächspartner (L) hat sodann die Aufgabe Hinweise für sein Verstehen der Präsentation zu geben.
- Streng genommen selbst wieder eine Presentation.

Wie stark muss der Hinweis des Verstehens sein, bis eine Contribution als grounded angesehen wird?
Dies ist der Fall, wenn das sogenannte Grounding-Kriterium erfüllt ist:
- Eine Contribution kann als grounded angenommen werden, wenn es zwischen S und L mutual knowledge ist, dass L verstanden hat was S meinte, sodass es sufficient for current purposes ist.

Evidence for understanding: 
Der Dialogpartner kann...
1. weiter Aufmerksam sein   
2. eine relevante nächste Äußerung machen
3. die Presentation explizit acknowledgen (z.B. durch Feedback, „mhm“)
4. demonstrieren, dass die Presentations verstanden wurde
5. paraphrasieren oder verbatim zeigen, dass die Presentation verstanden wurde

ist von schwach nach stark sortiert

Feedback Signale in der Kommunikation: 

- Verbale/vokale Signale, z.B.: ja, okay, mhm, m, hm, ach, oh, ...
- Stimmqualität, Längung, Kürzung, Duplizierung
- Kopfgestik, Gesichtsausdruck, Blick usw. 

Feedback als Evidence of understanding 

1. Setzt nicht voraus, den „Turn“ (das Rederecht) zu haben oder zu nehmen
2. Nicht störend/aufdringlich, keine Unterbrechung des Sprechers
3. Expressiv
4. Signalisiert bzw. zeigt den mentalen Zustand des Hörers an
5. Wird vom Sprecher benötigt und sogar über Hinweise/Signale „eingefordert“

Attentive Speaker-Systeme

Interaktive Systeme, die kommunikatives Feedback verarbeiten, um den Sprachdialog mit Nutzern zu verbessern
- Feedback/Backchannel-Signale erkennen und deuten
- Inferenzen über Hörerzustand der Nutzerin
- Feedback elizitieren, wenn nötig (z.B. unter Unsicherheit)
- Anpassung des eigenen Verhaltens (inkrementell, online und in Echtzeit)
- Anpassung über die Zeit (auch antizipativ)
